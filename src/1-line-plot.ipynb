### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt      # Step 1: Import Matplotlib  
import numpy as np                   # Import NumPy for numerical operations  

# Step 2: Define the x values  
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10  

# Step 3: Calculate the y values (y = x²)  
y = x ** 2                     # y values as the square of x  

# Step 4: Create the line plot  
plt.plot(x, y, label='y = x²')  # Plot x vs y with a label  

# Step 5: Customize the plot  
plt.title("Line Plot of y = x²")  # Add title  
plt.xlabel("X values")            # Label for the x-axis  
plt.ylabel("Y values (x squared)") # Label for the y-axis  
plt.grid(True)                    # Enable grid  
plt.legend()                      # Show legend  

# Step 6: Display the plot  
plt.show()                        # Render the plot
```
