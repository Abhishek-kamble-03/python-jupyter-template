### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt  # Step 1: Import Matplotlib  

# Step 2: Define the data  
labels = ['Python', 'Java', 'C++', 'JavaScript']  # Labels for the pie slices  
sizes = [40, 25, 20, 15]                          # Corresponding sizes for each label  

# Step 3: Create the pie chart  
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)  # Create the pie chart  

# Step 4: Customize the chart  
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle  
plt.title("Programming Language Usage Distribution")  # Optionally, add a title  

# Step 5: Display the plot  
plt.show()         # Render the chart
```
