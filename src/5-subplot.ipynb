### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt  
import numpy as np  

# Step 2: Define the data for each plot  
# Data for line chart  
x_line = np.linspace(0, 10, 100)  
y_line = np.sin(x_line)  

# Data for bar chart  
categories = ['A', 'B', 'C', 'D']  
values = [5, 7, 3, 9]  

# Data for scatter plot  
x_scatter = [1, 2, 3, 4, 5]  
y_scatter = [2, 4, 6, 8, 10]  

# Data for pie chart  
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']  
sizes_pie = [40, 25, 20, 15]  

# Step 3: Create a 2x2 subplot layout  
fig, axs = plt.subplots(2, 2, figsize=(10, 10))  

# Step 4: Plot each chart in its respective subplot  
# Line Chart  
axs[0, 0].plot(x_line, y_line, color='blue')  
axs[0, 0].set_title('Line Chart: y = sin(x)')  

# Bar Chart  
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])  
axs[0, 1].set_title('Bar Chart')  
axs[0, 1].set_ylabel('Values')  

# Scatter Plot  
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')  
axs[1, 0].set_title('Scatter Plot')  
axs[1, 0].set_xlabel('X values')  
axs[1, 0].set_ylabel('Y values')  

# Pie Chart  
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)  
axs[1, 1].set_title('Pie Chart')  

# Step 6: Optimize layout and display the plots  
plt.tight_layout()  # Adjust layout to prevent overlaps  
plt.show()          # Render the layout
```
